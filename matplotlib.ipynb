```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
houseprice=pd.read_csv('data/houseprice.csv',index_col=0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(houseprice['YrSold'],houseprice['SalePrice'])
plt.xlabel('Year', fontsize=5)
plt.ylabel('Price', fontsize=5)
plt.title('House Price based on Year')

plt.subplot(1, 2, 2)
plt.bar(houseprice['YearBuilt'],houseprice['SalePrice'])
plt.xlabel('Year Built', fontsize=5)
plt.ylabel('Sale Price', fontsize=5)
plt.title('Sale Price based on Built Year')

plt.tight_layout()
plt.show()
```
